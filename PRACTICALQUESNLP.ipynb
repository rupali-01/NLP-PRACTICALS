```python
# ===============================
# 1. Text Preprocessing
# ===============================
# Objective: Tokenize a paragraph, remove stopwords, apply stemming and lemmatization.

import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Example text
text = "The quick brown foxes were jumping over the lazy dogs in the park."

# Tokenization
tokens = word_tokenize(text)

# Stopword removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Stemming
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in filtered_tokens]

# Lemmatization using spaCy
nlp = spacy.load("en_core_web_sm")
doc = nlp(" ".join(filtered_tokens))
lemmatized = [token.lemma_ for token in doc]

print("Original Tokens:", tokens)
print("Filtered Tokens (Stopwords removed):", filtered_tokens)
print("Stemmed Tokens:", stemmed)
print("Lemmatized Tokens:", lemmatized)
```

```python
# ===============================
# 2. Bag of Words and TF-IDF Vectorization
# ===============================
# Objective: Convert text into numeric form using BoW and TF-IDF

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Example corpus
corpus = [
    "I love this movie. It's fantastic!",
    "This movie was terrible and boring.",
    "What a great movie! I’ll watch it again.",
]

# Bag of Words (BoW)
bow_vectorizer = CountVectorizer()
X_bow = bow_vectorizer.fit_transform(corpus)
print("\nBag of Words (BoW):\n", X_bow.toarray())

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(corpus)
print("\nTF-IDF Vectorization:\n", X_tfidf.toarray())
```

```python
# ===============================
# 3. Sentiment Analysis
# ===============================
# Objective: Classify text as positive, negative, or neutral.

from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Example texts
texts = [
    "I love this product! It's amazing.",
    "This is the worst experience I've had. I'm so disappointed.",
    "It's okay, not great, but not bad either."
]

# Sentiment Analysis using TextBlob
print("Sentiment Analysis using TextBlob:")
for text in texts:
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity  # -1 (negative) to 1 (positive)
    if sentiment > 0:
        print(f"Text: {text} | Sentiment: Positive")
    elif sentiment < 0:
        print(f"Text: {text} | Sentiment: Negative")
    else:
        print(f"Text: {text} | Sentiment: Neutral")

# Sentiment Analysis using VADER
analyzer = SentimentIntensityAnalyzer()
print("\nSentiment Analysis using VADER:")
for text in texts:
    score = analyzer.polarity_scores(text)
    if score['compound'] > 0:
        print(f"Text: {text} | Sentiment: Positive")
    elif score['compound'] < 0:
        print(f"Text: {text} | Sentiment: Negative")
    else:
        print(f"Text: {text} | Sentiment: Neutral")
```

```python
# ===============================
# 4. Named Entity Recognition (NER)
# ===============================
# Objective: Identify entities like names, places, and dates.

import spacy

# Load the spaCy pre-trained model
nlp = spacy.load("en_core_web_sm")

# Example text
text = """Apple is looking to buy a startup in the UK for $1 billion.
          Tim Cook, the CEO of Apple, is expected to meet the company's founder next week."""

# Process the text with spaCy
doc = nlp(text)

# Extract entities
print("Named Entities, Phrases, and Concepts:")
for entity in doc.ents:
    print(f"{entity.text} - {entity.label_}")

```

```python
# ===============================
# 5. Part-of-Speech (POS) Tagging
# ===============================
# Objective: Label words with their grammatical tags.

import spacy

# Load the spaCy pre-trained model
nlp = spacy.load("en_core_web_sm")

# Example text (from a famous book or article)
text = """The quick brown fox jumps over the lazy dog.
          It was an exciting moment for the team."""

# Process the text with spaCy
doc = nlp(text)

# Extract POS tags and count them
pos_tags = {}
for token in doc:
    if token.pos_ in pos_tags:
        pos_tags[token.pos_] += 1
    else:
        pos_tags[token.pos_] = 1

# Print POS tags for each token
print("POS Tags for Each Token:")
for token in doc:
    print(f"{token.text}: {token.pos_}")

# Print counts of each POS tag
print("\nPOS Tag Counts:")
for pos, count in pos_tags.items():
    print(f"{pos}: {count}")
```

```python
# ===============================
# 6. Text Classification
# ===============================
# Objective: Build a classifier to categorize text documents (e.g., spam vs. non-spam emails).

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Example dataset (spam vs. non-spam)
data = [
    ("Free money!!! Claim now!", "spam"),
    ("Your order has been shipped", "non-spam"),
    ("Earn cash quickly, no work required", "spam"),
    ("Hello, I hope you're doing well. Let's catch up soon.", "non-spam"),
    ("Win a free iPhone, just click here", "spam"),
    ("Please find the attached report for your review.", "non-spam"),
]

# Separate the text and labels
texts, labels = zip(*data)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.3, random_state=42)

# Convert text to numerical features using CountVectorizer
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train a Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train_vec, y_train)

# Predict the categories of the test set
y_pred = classifier.predict(X_test_vec)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Print predicted labels
print("\nPredictions:")
for text, label in zip(X_test, y_pred):
    print(f"Text: {text} | Predicted: {label}")
```

```python
# ===============================
# 7. Word Embeddings (Word2Vec)
# ===============================
# Objective: Represent words in vector space and find similar words.

import gensim.downloader as api

# Load pre-trained Word2Vec model (Gensim)
word_vectors = api.load("word2vec-google-news-300")

# Example words
words = ["king", "queen", "man", "woman", "apple", "banana"]

# Find similar words
print("Similar words for each word in the list:")
for word in words:
    similar_words = word_vectors.most_similar(word, topn=3)
    print(f"\n{word}:")
    for similar_word, score in similar_words:
        print(f"  {similar_word} (Similarity Score: {score:.3f})")
```

```python
# ===============================
# 8. Topic Modeling
# ===============================
# Objective: Extract topics from a collection of documents.

import gensim
from gensim import corpora
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Sample data: collection of documents (research abstracts)
documents = [
    "Machine learning is a field of artificial intelligence that focuses on building systems that can learn from data.",
    "Deep learning is a subset of machine learning, based on neural networks, which performs exceptionally well on complex tasks.",
    "Natural language processing deals with the interaction between computers and human languages.",
    "Computer vision involves teaching machines to interpret and understand the visual world."
]

# Preprocess the documents (tokenization and stopword removal)
stop_words = set(stopwords.words("english"))
processed_docs = []

for doc in documents:
    tokens = word_tokenize(doc.lower())  # Tokenize and convert to lowercase
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]  # Remove stopwords and punctuation
    processed_docs.append(tokens)

# Create a dictionary and a corpus
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

# Apply LDA (Latent Dirichlet Allocation) model
lda = gensim.models.LdaMulticore(corpus, num_topics=2, id2word=dictionary, passes=10)

# Print the topics extracted
print("Topics Extracted:")
for topic_id, topic in lda.print_topics(num_words=5):
    print(f"Topic {topic_id}: {topic}")
```

```python
# 9. Text Summarization
# Objective: Generate a summary of long text.
# Libraries: sumy, transformers
# Example Task: Summarize a Wikipedia article or blog post.
# Reinstalling/upgrading necessary libraries
!pip install --upgrade numpy
!pip install --upgrade transformers

# Install sumy if not already installed
!pip install sumy

# Import libraries
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from transformers import pipeline

# Function for extractive summarization
def extractive_summary(text, num_sentences=3):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LsaSummarizer()
    summary = summarizer(parser.document, num_sentences)
    return ' '.join(str(sentence) for sentence in summary)

# Function for abstractive summarization
def abstractive_summary(text):
    summarizer = pipeline("summarization")
    summary = summarizer(text, max_length=130, min_length=30, do_sample=False)
    return summary[0]['summary_text']

# Example long text (you can replace this with any long text)
long_text = """
Artificial intelligence (AI) is intelligence demonstrated by machines, in contrast to the natural intelligence displayed by humans and animals. Leading AI textbooks define the field as the study of "intelligent agents": any device that perceives its environment and takes actions that maximize its chance of successfully achieving its goals. Colloquially, the term "artificial intelligence" is often used to describe machines (or computers) that mimic "cognitive" functions that humans associate with the human mind, such as "learning" and "problem-solving".
"""

# Extractive Summary
print("Extractive Summary:")
print(extractive_summary(long_text, num_sentences=2))

# Abstractive Summary
print("\nAbstractive Summary:")
print(abstractive_summary(long_text))
```

```python
# 10. Text Classification
# Objective: Classify text into predefined categories.
# Libraries: scikit-learn, transformers
# Example Task: Classify a given text into categories such as 'Sports', 'Politics', 'Technology', etc.

# Install necessary libraries
!pip install --upgrade scikit-learn
!pip install --upgrade transformers

# Import libraries
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from transformers import pipeline

# Example dataset: Some sample text data and labels (categories)
texts = [
    "Football is a very popular sport worldwide.",
    "The latest advances in quantum computing are fascinating.",
    "The president gave a speech on national security today.",
    "Machine learning algorithms are revolutionizing technology.",
    "The stock market saw a significant rise today.",
    "Tennis players are preparing for the upcoming Wimbledon tournament."
]
labels = ["Sports", "Technology", "Politics", "Technology", "Finance", "Sports"]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.3, random_state=42)

# Create a text classification pipeline
model = make_pipeline(CountVectorizer(), MultinomialNB())

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test data
predictions = model.predict(X_test)

# Print predictions
print("Predictions:", predictions)

# Evaluate the model's performance
accuracy = model.score(X_test, y_test)
print("\nModel Accuracy:", accuracy)

# Example of using HuggingFace transformer for classification task
classifier = pipeline("zero-shot-classification")

# Classify an example sentence with zero-shot classification
text = "I just bought a new laptop with a high-end processor."
labels = ['Sports', 'Technology', 'Politics', 'Business']
result = classifier(text, candidate_labels=labels)

print("\nZero-Shot Classification Result:")
print(result)
```
